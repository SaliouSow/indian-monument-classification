```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
from PIL import Image
```

```python
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = Image.open(os.path.join(folder,filename))
        if img is not None:
            images.append(np.array(img))
    return images
```

```python
monument_folder = 'D:/WORKSPACE/CNN_Project/Dataset/Indian-monuments/images/train/tajmahal'
#print(monument_folder)
monument_images = load_images_from_folder(monument_folder)
```

```python
def load_data_from_folder(main_folder):
    all_data = []
    labels = []
    for idx, monument_folder in enumerate(os.listdir(main_folder)):
        monument_images = load_images_from_folder(os.path.join(main_folder, monument_folder))
        all_data += monument_images
        labels += [idx]*len(monument_images)
    return all_data, labels
```

```python
train_folder = 'path_to_your_train_folder'
train_data, train_labels = load_data_from_folder(train_folder)
```

```python
test_folder = 'path_to_your_test_folder'
test_data, test_labels = load_data_from_folder(test_folder)
```
